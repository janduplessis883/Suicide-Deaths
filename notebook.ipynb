# Suicide Death (Country and Year)

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/work/share-deaths-suicide.csv')
df.head()
```

```
df.tail(100)
```

```
df.rename(columns = {'Deaths - Self-harm - Sex: Both - Age: All Ages (Percent)':'Deaths'}, inplace = True)
```

```
df.info()
```

```
df.columns
```

```
df.Entity.nunique()
```

```
df[    
    (df["Code"] == "AFG") &    
    (df["Year"] > 2000)
].groupby(["Code", "Year"])["Deaths"].max().plot(x='Year', y='Deaths', label='AFG')
df[    
    (df["Code"] == "ZWE") &    
    (df["Year"] > 2000)
].groupby(["Code", "Year"])["Deaths"].max().plot(x='Year', y='Deaths', label="ZWE")
```

```
df.Code.value_counts()
```

```
import numpy as np
list_of_codes = df.Code.to_list()
list_of_codes = np.unique(list_of_codes)
print(list_of_codes)
```

```
import numpy as np
list_of_entity = df.Entity.to_list()
list_of_entity = np.unique(list_of_entity)
print(list_of_entity)
```

```
df[        
    (df["Year"] > 1990)
    ].groupby(["Code", "Year"])["Deaths"].max().plot()
```

```
df[        
    (df["Year"] > 1990)
    ].groupby(["Code", "Year"])["Deaths"].mean().plot()
```

```
for c in list_of_codes:
    df[    
        (df["Code"] == c) &    
        (df["Year"] > 1990)
    ].groupby(["Code", "Year"])["Deaths"].max().plot()
```

```
for c in list_of_codes:
    df[    
        (df["Code"] == c) &    
        (df["Year"] > 1990)
    ].groupby(["Code", "Year"])["Deaths"].max().plot()

    print(c)
    plt.show()
```
